```python
eda_data = eda_data.drop('names', axis=1)
eda_data = drop_duplicates(eda_data, True)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="most_frequent")
eda_data[["transmission"]] = imputer.fit_transform(eda_data[["transmission"]]) # Theo lý thuyết thì đa số dòng xe hiện tại sẽ sử dụng hộp số tự động nên chỉ cần fill tự động theo mode là đủ
```

```python
# HANDLE FUEL AND VOULUME 
eda_data["fuel"] = eda_data["fuel"].fillna(eda_data.groupby(["brand", "model"])["fuel"].transform(lambda x: x.mode().iloc[0])) # Theo lý thuyết đây là dữ liệu khuyết thiếu MCR, ta có thể suy luận rầng loại nhiên liệu chúng xài sẽ tương tự những mẫu của nó có sẵn
eda_data["volumne"] = eda_data["volumne"].fillna(eda_data.groupby(["brand", "model"])["volumne"].transform("mean")) # Theo lý thuyết đây cũng là dữ liệu khuyết thiếu MCR, nhưng ta có thể suy luận rằng dung tích xy lanh nó cũng sẽ trùng với những mẫu xe cùng dòng
eda_data["volumne"] = eda_data["volumne"].fillna(0) # Xử lý đối với loại xe điện
```
